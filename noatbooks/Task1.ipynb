```python
num_col= data.select_dtypes(include=['number'])
var=num_col.var()
# Calculate Coefficient of Variation
cv = num_col.std() / num_col.mean()
variation_table=pd.DataFrame({
    "Variance": var,
    "Coefficient of Variation":cv
})
variation_table
```

```python
num_col = num_col.drop('NumberOfVehiclesInFleet', axis=1)

```

```python
data_mode=data.mode()

data_mode
```

```python
import matplotlib.pyplot as plt
for col in num_col:
    plt.figure(figsize=(8, 6))  
    plt.hist(data[col], bins=30, color='blue', edgecolor='black') 
    plt.title(f"Histogram of {col}") 
    plt.xlabel(col)  
    plt.ylabel("Frequency") 
    plt.grid(True)  
    plt.show()
```

```python
cat_col = data.select_dtypes(include=['object'])
cat_col = cat_col.drop('CapitalOutstanding', axis=1)
```

```python


for col in cat_col:
    plt.figure(figsize=(8, 6))
    # Count occurrences of each category
    counts = data[col].value_counts()
    
    # Ensure the index is a list and values are a numpy array
    categories = counts.index.tolist()
    values = counts.values
    
    plt.barh(categories, values, color='purple', edgecolor='black')
    plt.title(f'Bar Chart for {col}')  
    plt.xlabel('Count') 
    plt.ylabel(col)  
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()

```
