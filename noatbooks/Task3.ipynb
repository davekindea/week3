```python
import pandas as pd
data=pd.read_csv("./Data/MachineLearningRating_v3.txt",sep="|")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
categorical_columns = data.select_dtypes(include=['object']).columns
for column in categorical_columns:
    data[column] = data[column].fillna(data[column].mode()[0])
num_col=data.select_dtypes(include=['float64', 'int64']).columns
for col in num_col:
    data[col] = data[col].fillna(data[col].mean())
data.isnull().sum()
```

```python
data["NumberOfVehiclesInFleet"]=data["NumberOfVehiclesInFleet"].fillna(0)
```

```python

from scipy.stats import chi2_contingency

province_group = data.groupby('Province').agg({'TotalClaims': 'sum', 'SumInsured': 'sum'}).reset_index()
province_group['ClaimsRatio'] = province_group['TotalClaims'] / province_group['SumInsured']


contingency_table = pd.crosstab(data['Province'], data['TotalClaims'])

chi2, p_value, dof, expected = chi2_contingency(contingency_table)


res = pd.DataFrame(
    {
        "Chi-Square Test statistic": [chi2],
        "p_value": [p_value],
        "dof": [dof],
        "expected": [expected]
    }
)

if p_value > 0.05:
    res["status"] = "accepted"
else:
    res["status"] = "rejected"
contingency_table 
```

```python
Postalcode_group = data.groupby('PostalCode').agg({'TotalClaims': 'sum', 'SumInsured': 'sum'}).reset_index()


Postalcode_group['ClaimsRatio'] = Postalcode_group['TotalClaims'] / Postalcode_group['SumInsured']


contingency_table_zip = pd.crosstab(data['PostalCode'], data['TotalClaims'])


chi2_zip, p_value_zip, dof_zip, expected_zip = chi2_contingency(contingency_table_zip)
contingency_table_zip
```

```python
res_zip = pd.DataFrame(
    {
        "Chi-Square Test statistic for postalCode": [chi2_zip],
        "p_value postalCode": [p_value_zip],
        "dof postalCode": [dof_zip],
        "expected postalCode": [expected_zip]
    }
)

if p_value > 0.05:
    res_zip["status postalCode"] = "accepted"
else:
    res_zip["status postalCode"] = "rejected"
res_zip
```

```python
from scipy.stats import f_oneway
data['ProfitMargin'] = (data['TotalPremium'] - data['TotalClaims']) / data['TotalPremium']
zipcode_profit_group = data.groupby('PostalCode')['ProfitMargin'].mean().reset_index()


profit_margins_by_zip = [group['ProfitMargin'].values for name, group in data.groupby('PostalCode')]

anova_stat, anova_p_value = f_oneway(*profit_margins_by_zip)
profit_margins_by_zip

```

```python
res_ProfitMargin=pd.DataFrame({
    "'ANOVA Test statistic": [anova_stat],
    "p-value:": [anova_p_value]

})
if anova_p_value >= 0.05:
    res_ProfitMargin["status"]="Accept"
else:
    res_ProfitMargin["status"]="Reject"
res_ProfitMargin
```

```python
from scipy.stats import ttest_ind

claims_men = data[data['Gender'] == 'Male']['TotalClaims'] / data[data['Gender'] == 'Male']['SumInsured']
claims_women = data[data['Gender'] == 'Female']['TotalClaims'] / data[data['Gender'] == 'Female']['SumInsured']

t_stat, p_value_gender = ttest_ind(claims_men, claims_women)


res_claims=pd.DataFrame ({
    "T-Test statistic" : [t_stat],
    "p-value" : [p_value_gender]

})

if p_value_gender < 0.05:
    res_claims["Status"]= "Reject"
else:
    res_claims["Status"]="Accept"
res_claims
```
